```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from shapely.wkt import loads

# 주어진 문자열을 파싱하여 Shapely의 Polygon 객체로 변환
polygon_wkt = df_all['폴리곤_좌표'][0]
polygon = loads(polygon_wkt)

# Polygon 객체에서 좌표를 추출
polygon_coords = list(polygon.exterior.coords)

# 시각화
fig, ax = plt.subplots()
polygon_patch = Polygon(polygon_coords, closed=True, edgecolor='r', facecolor='none')
ax.add_patch(polygon_patch)

# X, Y 좌표 범위 설정 (원하는 범위로 조절)
ax.set_xlim(200000, 205000)  # 예시 범위, 필요에 따라 조절
ax.set_ylim(443500, 449000)  # 예시 범위, 필요에 따라 조절

# 가로 세로 비율 조정
plt.gca().set_aspect('equal', adjustable='box')

# 그래프 표시
plt.show()
```

```python
from shapely.geometry import Polygon, Point
import shapely.wkt

# 주어진 폴리곤 좌표 문자열을 파싱하여 Shapely Polygon 객체 생성
polygon_coords_str = df_all['폴리곤_좌표'][0]  # 문자열 가져오기
polygon = shapely.wkt.loads(polygon_coords_str)  # Shapely Polygon 객체로 변환

# 검사할 점들을 Shapely Point 객체로 생성한 리스트
point_to_check_list = [Point(201700, 446800), Point(201800, 446700)]

# 각 점에 대한 결과를 저장할 리스트 초기화
results = []

# 각 점을 폴리곤 내에 있는지 확인하고 결과를 리스트에 추가
for point in point_to_check_list:
    is_inside = polygon.contains(point)
    results.append(is_inside)

# 결과 출력
for i, is_inside in enumerate(results):
    if is_inside:
        print(f"점 {i+1}은 폴리곤 내에 있습니다.")
    else:
        print(f"점 {i+1}은 폴리곤 내에 없습니다.")
```

```python
from shapely.wkt import loads
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, Point
import shapely.wkt
import geopandas as gpd

# Assuming you have loaded your polygon data into df_all['폴리곤_좌표'][0]

# Create a Shapely Polygon object from the WKT string
polygon_coords_str = df_all['폴리곤_좌표'][0]
polygon = shapely.wkt.loads(polygon_coords_str)

# Create a GeoDataFrame for visualization
gdf = gpd.GeoDataFrame(geometry=[polygon])

# Create a list of points to check
point_to_check_list = [Point(201700, 446800), Point(201800, 446700)]

# Initialize a list to store the results
results = []

# Check if each point is inside the polygon and add the results to the list
for point in point_to_check_list:
    is_inside = polygon.contains(point)
    results.append(is_inside)

# Visualization
fig, ax = plt.subplots()

# Plot the polygon
gdf.plot(ax=ax, alpha=0.5, color='blue', edgecolor='black')

# Plot the points
for i, point in enumerate(point_to_check_list):
    ax.plot(point.x, point.y, 'ro', label=f'Point {i + 1}')

# Add labels to the points
for i, point in enumerate(point_to_check_list):
    ax.annotate(f'Point {i + 1}', (point.x, point.y), textcoords="offset points", xytext=(0,10), ha='center')

# Set axis limits based on polygon coordinates
min_x, min_y, max_x, max_y = polygon.bounds
ax.set_xlim(min_x - 100, max_x + 100)
ax.set_ylim(min_y - 100, max_y + 100)

# Display the plot
plt.legend()
plt.show()

# Print the results
for i, is_inside in enumerate(results):
    if is_inside:
        print(f"점 {i + 1}은 폴리곤 내에 있습니다.")
    else:
        print(f"점 {i + 1}은 폴리곤 내에 없습니다.")
```
